```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
plt.rcParams['font.size'] = 20
mpl.rcParams['xtick.labelsize'] = 20
mpl.rcParams['ytick.labelsize'] = 20 
mpl.rcParams['axes.labelsize'] = 20
mpl.rcParams['ytick.major.size'] = 15
mpl.rcParams['xtick.major.size'] = 15
```

```python
path = r"D:\progs\fuzzy-chainsaw\2024_2025\IRUSIA\data\API_NY.GDP.MKTP.CD_DS2_en_csv_v2_120252.csv"

df = pd.read_csv(path, comment='#', header=None)

df
```

```python
df_rus = df[df[0] == "Russian Federation"]

df_header = df[df[0] == "Country Name"]

df_header.reset_index(drop=True, inplace=True)
df_rus.reset_index(drop=True, inplace=True)

df_rus
```

```python
new_df_rus = {
    'year' : df_header.iloc[0][4:].values.flatten(),
    'GDP' : df_rus.iloc[0][4:].values.flatten()
}

new_df_rus = pd.DataFrame.from_records(new_df_rus)
new_df_rus = new_df_rus.query("year > 2012 & year < 2024")

new_df_rus.reset_index(drop=True, inplace=True)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8), tight_layout=True)

ax.bar(new_df_rus['year'], new_df_rus['GDP']/1e12, color='lightslategrey')
# ax.plot(new_df_rus['year'], new_df_rus['GDP']/1e12, color='lightslategrey')
# ax.scatter(new_df_rus['year'], new_df_rus['GDP']/1e12, color='black', marker='.', s=100)
ax.set_title("GDP of russian federation, 2013-2023")
ax.tick_params(top=False, labeltop=False, bottom=False, labelbottom=True, right=True, direction='in', labelsize=15)

ax.set_ylabel("GDP, trillions USD")
ax.set_xlabel("Year")
ax.set_yticks(np.arange(0.0, 2.5, 0.2))
ax.set_xticks(np.arange(2013, 2024, 1))

array = np.arange(2013, 2024, 1)

labels = array.tolist()
ax.set_xticklabels(labels, rotation=45, ha='center')

rise_fall = []
for i, gdp in enumerate(new_df_rus['GDP'].array[1:]/1e12):
    print(i, gdp)
    rise_fall.append((gdp/(new_df_rus['GDP'][i]/1e12) - 1)*100)

print(rise_fall)

for i, rf in enumerate(rise_fall):
    if rf < 0:
        ax.text(new_df_rus['year'][i+1], new_df_rus['GDP'][i+1]/1e12 + 0.05, f"{round(rf, 1)}" + r"%", ha='center', va='center', fontsize=12)
    else:
        ax.text(new_df_rus['year'][i+1], new_df_rus['GDP'][i+1]/1e12 + 0.05, f"+{round(rf, 1)}" + r"%", ha='center', va='center', fontsize=12)

plt.show()
```
