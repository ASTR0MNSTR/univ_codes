```python
path_to_stata = r"D:\progs\fuzzy-chainsaw\2024_2025\IRUSIA\sanctions\data\GSDB_V4_dyadic.dta"

stata = pd.read_stata(path_to_stata)

stata
```

```python
stata_rf = stata.query("sanctioned_state == 'Russia'")

stata_rf.reset_index(drop=True, inplace=True)

years = [] 
for i, row in stata_rf.iterrows():
    years.append(int(row['year'].strftime('%Y-%m-%d').split("-")[0]))

stata_rf['pure_years'] = years

stata_rf = stata_rf[stata_rf['pure_years'] >= 2022]

EU_countries = ['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Republic of Cyprus', 'Czech Republic', 'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Ireland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia', 'Spain', 'Sweden']

stata_rf = stata_rf[stata_rf['sanctioning_state'].isin(EU_countries)]

stata_rf
```

```python
stata_rf.to_csv(r"D:\progs\fuzzy-chainsaw\2024_2025\IRUSIA\sanctions\data\after_cleaning.csv", index=False)
```
