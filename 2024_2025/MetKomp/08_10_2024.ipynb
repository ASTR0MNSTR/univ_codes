```python
from scipy.optimize import curve_fit 
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0.65
b = 10.3
x = np.arange(10, 100)
y = a*x + b + 3*np.random.random(len(x))

data = np.column_stack([x, y])
np.savetxt(r"./DATA/dane.dat", data)

plt.style.use("fivethirtyeight")
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

x, y = np.loadtxt(r"E:\progs\fuzzy-chainsaw\2024_2025\MetKomp\DATA\dane.dat", unpack=True)

def line(x, a, b):
    return a*x + b

popt, popc = curve_fit(line, x, y)

a = popt[0]
b = popt[1]

print(a, b)
plt.scatter(x, y)
plt.plot(x, line(x, a, b), color='r')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
import os

def line(x, a, b):
    return a*x + b
func_list = [line, line, line, line, line, line, line, line, line]

for i, file in enumerate(os.listdir(r"E:\progs\fuzzy-chainsaw\2024_2025\MetKomp\DATA")):
    x, y = np.loadtxt(r"E:\progs\fuzzy-chainsaw\2024_2025\MetKomp\DATA\{}".format(file), unpack=True)
    popt, popc = curve_fit(func_list[i], x, y)
    plt.scatter(x, y)
    plt.plot(x, func_list[i](x, *popt), color='r')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(file)
    plt.show()
```
