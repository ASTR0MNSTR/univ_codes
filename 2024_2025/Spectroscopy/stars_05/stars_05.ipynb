```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path_to_template = r"D:\progs\fuzzy-chainsaw\2024_2025\Spectroscopy\stars_05\DATA\template.txt"

temp = pd.read_csv(path_to_template, sep=r'\s+', header=None, names=['REF', 'NAME'])

# temp

ref_name_dict = dict(zip(temp['REF'], temp['NAME']))

ref_name_dict
```

```python
path_to_res = r"D:\progs\fuzzy-chainsaw\2024_2025\Spectroscopy\stars_05\DATA\cros_1.txt"
res = pd.read_csv(path_to_res, sep="\s+", header=None, names=["OBJECT", "IMAGE", "REF", "HJD", "AP", "CODES", "SHIFT", "HGHT", "FWHM", "TDR", "VOBS", "VREL", "VHELIO", "VERR"])
res
```

```python
positions = np.arange(len(res))

# Plot
plt.figure(figsize=(16, 10))
plt.bar(positions, res['HGHT'], color='mediumseagreen')

# X-axis with REF names

tick_names = [ref_name_dict[key] for key in res['REF']]
plt.xticks(positions, tick_names, rotation=90, ha='center')  # Rotate if names are long
plt.xlabel('REF')
plt.ylabel('HGHT')
plt.title('star1')

plt.tight_layout()
plt.show()
```
