```python
import pandas as pd
from datetime import datetime
from astropy.time import Time

# Sample DataFrame with filenames
# df = pd.DataFrame({'filename': ['gamPeg_2011-09-29_18-57-54.fits', 'gamPeg_2011-09-29_19-10-01.fits']})

def extract_jd_from_filename(fname):
    try:
        # Remove extension and split
        parts = fname.rstrip('.fits').split('_')
        date_str = parts[1]  # '2011-09-29'
        time_str = parts[2].replace('-', ':')  # '18:57:54'

        # Combine into datetime
        dt = datetime.strptime(f"{date_str} {time_str}", "%Y-%m-%d %H:%M:%S")

        # Convert to JD
        t = Time(dt, scale='utc')
        return t.jd
    except Exception as e:
        print(f"Failed to parse {fname}: {e}")
        return None

# Assuming df is your existing DataFrame with a 'filename' column
df['JD'] = df[0].apply(extract_jd_from_filename)

# Show the result
print(df)
```

```python
plt.scatter(df['JD'], df[11], marker='.', s=25, color='blue')

plt.xlabel('JD')
plt.ylabel('RV, km/s')

plt.savefig('plot.pdf', dpi=600, format='pdf')
```
