```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
beta = np.arange(1, 500, 1)
mu = np.arange(0, 1, 0.01)

def mean_R(beta, mu):
    top = 1 / (np.exp(beta * (1 - mu)) - 1)
    bottom = (np.exp(-1 * beta * (mu + 2)))/(1 - np.exp(-1*beta)) + top

    return top/bottom

from mpl_toolkits.mplot3d import axes3d

ax = plt.figure().add_subplot(projection='3d')

X, Y = np.meshgrid(beta, mu)
Z = mean_R(X, Y)

# Plot the 3D surface
ax.plot_surface(X, Y, Z, edgecolor='royalblue', lw=0.5, rstride=8, cstride=8,
                alpha=0.3)

# percentiles = [5, 25, 50, 75, 95]
# levels = np.percentile(Z, percentiles)

# ax.contour(X, Y, Z, levels=levels, zdir="z", offset=Z.min(), cmap="plasma")
# ax.contour(X, Y, Z, levels=levels, zdir="x", offset=X.min(), cmap="plasma")
# ax.contour(X, Y, Z, levels=levels, zdir="y", offset=Y.min(), cmap="plasma")

ax.set(xlim=(1, 500), ylim=(0, 1), zlim=(0.9, 1.1),
       xlabel=r'$\mathrm{\beta}$', ylabel=r'$\mathrm{\mu}$', zlabel=r'$\mathrm{R}$')

plt.show()
```
