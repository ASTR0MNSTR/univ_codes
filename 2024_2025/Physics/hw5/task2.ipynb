```python
import matplotlib.pyplot as plt
import numpy as np

spin = 0.5
B = 1 #Tesla
g = 2.0                   # mgnetic constant
mu_B = 9.2740100783e-24   # J/T, magneton Bohr
k_B  = 1.380649e-23       # J/K
```

```python
Delta = g * mu_B * B      # J
Delta_over_kB = Delta / k_B

def P_ground(T):
    x = Delta_over_kB / T   # = Delta/(k_B T)
    return 1.0 / (1.0 + np.exp(-x))

Ts = np.concatenate((np.linspace(0.01,0.5,100), np.linspace(0.5,500,300)))
Ps = P_ground(Ts)

plt.figure(figsize=(7,4))
plt.plot(Ts, Ps)
plt.axvline(Delta_over_kB, color='gray', linestyle='--', label=f'Δ/kB ≈ {Delta_over_kB:.3f} K')
plt.xscale('log')
plt.xlabel('T (K)')
plt.ylabel('P_ground')
plt.title('Probability of ground state vs. T (B=1 T)')
plt.legend()
plt.grid(True)
plt.show()
```
