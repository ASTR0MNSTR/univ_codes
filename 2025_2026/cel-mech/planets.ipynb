### Position 1

#### barycentrum determination

```python
xb_1 = np.sum(pos1['X']*mass['mass'])/(np.sum(mass['mass']) + 1.98e30)
yb_1 = np.sum(pos1['Y']*mass['mass'])/(np.sum(mass['mass']) + 1.98e30)
zb_1 = np.sum(pos1['Z']*mass['mass'])/(np.sum(mass['mass']) + 1.98e30)

print("CM at time 1: ", xb_1, yb_1, zb_1)
```

```python
vxb_1 = np.sum(pos1['VX']*mass['mass'])/(np.sum(mass['mass']) + 1.98e30)
vyb_1 = np.sum(pos1['VY']*mass['mass'])/(np.sum(mass['mass']) + 1.98e30)
vzb_1 = np.sum(pos1['VZ']*mass['mass'])/(np.sum(mass['mass']) + 1.98e30)

print("Velocity of the CM at time 1: ", vxb_1, vyb_1, vzb_1)
```

```python
rb = np.array([xb_1, yb_1, zb_1])
vb = np.array([vxb_1, vyb_1, vzb_1])

rb
```

```python
vb
```

#### Shifting to barycentrum coordinates 

```python
r1 = pos1[['X','Y','Z']].to_numpy()
v1 = pos1[['VX','VY','VZ']].to_numpy()

r1_cm = r1 - rb
v1_cm = v1 - vb

v1_cm
```

#### Angular momentum determinantion

```python
m = mass['mass'].to_numpy()   

L = m[:, None] * np.cross(r1_cm, v1_cm)

L_total = np.sum(L, axis=0) + 1.98e30 * np.cross(-rb, -vb) 

L_total
```

### Position 2

```python
xb_2 = np.sum(pos2['X']*mass['mass'])/(np.sum(mass['mass']) + 1.98e30)
yb_2 = np.sum(pos2['Y']*mass['mass'])/(np.sum(mass['mass']) + 1.98e30)
zb_2 = np.sum(pos2['Z']*mass['mass'])/(np.sum(mass['mass']) + 1.98e30)

print("CM at time 2: ", xb_2, yb_2, zb_2)
```

```python
vxb_2 = np.sum(pos2['VX']*mass['mass'])/(np.sum(mass['mass']) + 1.98e30)
vyb_2 = np.sum(pos2['VY']*mass['mass'])/(np.sum(mass['mass']) + 1.98e30)
vzb_2 = np.sum(pos2['VZ']*mass['mass'])/(np.sum(mass['mass']) + 1.98e30)

print("Velocity of the CM at time 2: ", vxb_2, vyb_2, vzb_2)
```

```python
rb2 = np.array([xb_2, yb_2, zb_2])
vb2 = np.array([vxb_2, vyb_2, vzb_2])

rb2
```

```python
r2 = pos2[['X','Y','Z']].to_numpy()
v2 = pos2[['VX','VY','VZ']].to_numpy()

r2_cm = r2 - rb2
v2_cm = v2 - vb2

v2_cm
```

```python
m = mass['mass'].to_numpy()   

L = m[:, None] * np.cross(r2_cm, v2_cm)

L_total2 = np.sum(L, axis=0) + 1.98e30 * np.cross(-rb2, -vb2) 

L_total2
```

```python
L_JUP = m[4] * np.cross(r2_cm[4, :], v2_cm[4, :])

L_JUP
```
