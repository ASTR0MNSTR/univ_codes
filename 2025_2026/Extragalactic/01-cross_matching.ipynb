### TASK 1: load LambdarCatv01.fits as Pandas DataFrame, see what's inside

```python
import pandas as pd
from astropy.io import fits
from astropy.table import Table
```

```python
with fits.open(lambdar_path) as hdul:
   data = hdul[1].data

data
```

```python
table_lambda_r = Table(data)

table_lambda_r
```

### TASK 2: load SersicCatSDSSv09.fits as Pandas DataFrame, see what's inside

```python
with fits.open(sersic_path) as hdul:
   data = hdul[1].data

data

table_sersic = Table(data)

table_sersic
```

### TASK 3: cross-match lambdar and sersic catalogues using RA and DEC (use `SkyCoord.match_to_catalog_sky()`)

```python
from astropy.coordinates import SkyCoord
from astropy import units as u
```

```python
lambdar_coords = SkyCoord(ra=table_lambda_r['RA']*u.degree, dec=table_lambda_r['DEC']*u.degree)

sersic_coords = SkyCoord(ra=table_sersic['RA']*u.degree, dec=table_sersic['DEC']*u.degree)

sersic_coords
```

```python
idx, d2d, _ = lambdar_coords.match_to_catalog_sky(sersic_coords)
```

```python
len(idx)
```

```python
d2d.max()
```

```python
max_sep = 0.5 * u.arcsec

max_sep
```

```python
d2d < max_sep
```

```python
len(idx[d2d < max_sep])
```

### TASK 4: cross-match lambdar and sersic catalogues using `CATAID` columnn

```python
sersic_df = table_sersic.to_pandas()

sersic_df
```

```python
lambdar_df = table_lambda_r.to_pandas()

lambdar_df
```

```python
df_mixed = pd.merge(lambdar_df, sersic_df, on='CATAID', how='right')

df_mixed
```
