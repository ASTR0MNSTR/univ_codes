```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
from astropy import units as u
import astropy.cosmology as cosmo
from astropy.cosmology import FlatLambdaCDM
cosm = FlatLambdaCDM(H0=70, Om0=0.3, Tcmb0=2.725)
```

```python
import numpy as np
```

### TASK 1: How Hubble's constant tells us about the age of the Universe?

#### 1a: calculate the age on the white/black-board

```python
((1/(70*u.km.to(u.Mpc)))*u.s).to(u.Gyr)  
```

### TASK 2: using some concordance flat $\Lambda$CDM model, calculate age at some redshifts and plot them.
#### try to "flatten" the passage of time -- what does it tell you about redshift?

```python
z = list(np.arange(0, 20, 1))
age = [(cosm.age(item)).value for item in z]

age
```

```python
fig, axs = plt.subplots(1, 1, figsize=(12, 12))

plt.scatter(z, age, color='magenta', marker='.')
plt.xlabel('redshift')
plt.ylabel('age / Gyr')

plt.tight_layout()
plt.show()
```
