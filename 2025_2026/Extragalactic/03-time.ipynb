### TASK 2: using some concordance flat $\Lambda$CDM model, calculate age at some redshifts and plot them.
#### try to "flatten" the passage of time -- what does it tell you about redshift?

```python
ages = np.arange(0.1, 10, 1)

# redshifts = z_at_value(cosm.age, 10 * u.Gyr)
redshifts = [z_at_value(cosm.age, age * u.Gyr) for age in ages]
```

```python
fig, axs = plt.subplots(1, 1, figsize=(6, 6))

plt.scatter(ages, redshifts, color='mediumvioletred', marker='.')
plt.xlabel('age / Gyr')
plt.ylabel('redshift')
plt.title('Redshift vs. Age of the Universe')
plt.tight_layout()
plt.show()
```
