```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tau = [0, 0.5, 1.0, 2.0, 3.0, 5.0, 10.0]
tau_array = np.array(tau)

plt.plot(np.exp(-tau_array), tau_array, 'o-')
# plt.yscale('log')
plt.ylabel('Optical Depth (tau)')
plt.xlabel('Intensity (I/I0)')
plt.title('Exponential Decay of Intensity with Optical Depth')
plt.grid(True)
plt.show()
```
