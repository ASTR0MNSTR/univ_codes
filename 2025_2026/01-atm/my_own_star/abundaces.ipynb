```python
import matplotlib.pyplot as plt
```

```python
element_wise_abundance = {
    'C' : [-3.72 + 12.0, 0.06],
    'Na' : [-5.88 + 12.0, 0.05],
    'Mg' : [-4.72 + 12.0, 0.03],
    'Si' : [-4.51 + 12.0, 0.04],
    'S' : [-4.87 + 12.0, 0.15],
    'Ca' : [-5.71 + 12.0, 0.02],
    'Ti' : [-7.13 + 12.0, 0.01],
    'V' : [-7.98 + 12.0, 0.09],
    'Cr' : [-6.54 + 12.0, 0.02],
    'Mn' : [-6.82 + 12.0, 0.04],
    'Fe' : [-4.71 + 12.0, 0.01],
    'Ni' : [-5.99 + 12.0, 0.02],
}

element_wise_sun = {
    'C' : [8.52, 0.06],
    'Na' : [6.33, 0.03],
    'Mg' : [7.58, 0.05],
    'Si' : [7.55, 0.05],
    'S' : [7.33, 0.05],
    'Ca' : [6.36, 0.02],
    'Ti' : [5.02, 0.06],
    'V' : [4.00, 0.02],
    'Cr' : [5.67, 0.03],
    'Mn' : [5.39, 0.03],
    'Fe' : [7.50, 0.05],
    'Ni' : [6.25, 0.04], 
}

element_wise_number = {
    'C' : 6,
    'Na' : 11,
    'Mg' : 12,
    'Si' : 14,
    'S' : 16,
    'Ca' : 20,
    'Ti' : 22,
    'V' : 23,
    'Cr' : 24,
    'Mn' : 25,
    'Fe' : 26,
    'Ni' : 28,
}
```

```python
fig, ax = plt.subplots(figsize=(12,6), tight_layout=True)

Y = []
nums = []

for key in element_wise_abundance.keys():
    ys = element_wise_abundance[key][0] - element_wise_sun[key][0]
    ax.errorbar(element_wise_number[key], ys, yerr=element_wise_abundance[key][1] + element_wise_sun[key][1], fmt='o', color='black')
    ax.text(element_wise_number[key]+0.1, ys+0.005, key, fontsize=12)
    Y.append(ys)
    nums.append(element_wise_number[key])

ax.plot(nums, Y, color='black', alpha = 0.5, linestyle='-.')
ax.axhline(0.0, color='black', linestyle='--', alpha=0.5)
ax.set_xlabel('Atomic Number', fontsize=14)
ax.set_ylabel(r'$\Delta A(E)$', fontsize=14)

plt.savefig('abundance_differences.pdf', dpi=600, bbox_inches='tight')
```
