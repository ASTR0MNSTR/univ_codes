```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('period_scan.log', comment='#', delim_whitespace=True, names=['period', 'rms', 'chi2', 'iter', 'dark_area_percent'], header=None)

df
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(df['period'], df['rms'], marker='o', linestyle='-', color='b')
ax.set_xlabel('Period')
ax.set_ylabel('RMS')
ax.set_title('RMS vs Period Scan')
ax.grid(True)
plt.show()
```

```python
best_period = df.loc[df['rms'].idxmin()]['period']
print(f"Best period with minimum RMS: {best_period}")
```
