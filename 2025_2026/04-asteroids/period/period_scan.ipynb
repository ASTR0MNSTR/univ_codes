```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('period_scan.log', comment='#', delim_whitespace=True, names=['period', 'rms', 'chi2', 'iter', 'dark_area_percent'], header=None)

df

print(np.diff(df['period']))
```

```python
best_period = df.loc[df['rms'].idxmin()]['period']
uncertainty = np.diff(df['period'])[df['rms'].idxmin()]
print(f"Best period with minimum RMS: {best_period}")
print(f"Uncertainty in period: {uncertainty}")

min_rms = 1.1*df['rms'].min()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(df['period'], df['rms'], marker='o', linestyle='-', color='b')
ax.set_xlabel('Period')
ax.set_ylabel('RMS')
ax.set_title('RMS vs Period Scan')
ax.grid(True)
ax.axhline(y=min_rms, color='r', linestyle='--', label='1.10 * Min RMS')
ax.legend()
plt.show()
```
