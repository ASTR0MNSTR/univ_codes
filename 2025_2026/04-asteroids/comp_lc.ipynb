```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
path = r'/media/olehryz/Elements/progs/univ_codes/2025_2026/04-asteroids/Z0859.ATL'
```

```python
observ_blocks = []
with open(path) as f:
    content = [item.strip() for item in f.readlines()]
    
    #block separator
    block = []
    for line in content:
        if line.startswith('=') == False and line.startswith('DATA') == False:
            block.append(line)
        else:
            observ_blocks.append(block)
            block = []

atl_headers = [item for i, item in enumerate(observ_blocks) if i % 2 == 0]
atl_data = [item for i, item in enumerate(observ_blocks) if i % 2 == 1]

atl_data
```

```python
import numpy as np

NewFrameDict = {
}

df = pd.DataFrame(NewFrameDict)

for j, block in enumerate(atl_data):
    JD_all = [line.split()[0] for line in block]
    print(JD_all)
    JD = np.array([float(jd) if jd[0] != '!' else float(jd[1:]) for jd in JD_all])
    flags = [True if jd[0] != '!' else False for jd in JD_all]
    mag = np.array([float(line.split()[1]) for line in block])
    # mag_err = np.array([float(line.split()[2]) for line in block])
    ind = np.array([i for i, item in enumerate(block)])
    ind_header = np.array([j for item in block])

    # print(flags)
    
    NewFrameDict = {
        'ind_header' : ind_header,
        'ind' : ind,
        'JD' : JD,
        'mag' : mag,
        # 'mag_err' : mag_err,
        'flags' : flags,
    }
    
    ndf = pd.DataFrame(NewFrameDict)
    
    df = pd.concat([df, ndf], axis = 0, join='outer')

df.info()
df.to_csv(r'/media/olehryz/Elements/progs/univ_codes/2025_2026/04-asteroids/atl.csv', index=False)
```
