```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
path = r'/media/olehryz/Elements/progs/univ_codes/2025_2026/04-asteroids/Z0859.ATL'
```

```python
observ_blocks = []
with open(path) as f:
    content = [item.strip() for item in f.readlines()]
    
    #block separator
    block = []
    for line in content:
        if line.startswith('=') == False and line.startswith('DATA') == False:
            block.append(line)
        else:
            observ_blocks.append(block)
            block = []

atl_headers = [item for i, item in enumerate(observ_blocks) if i % 2 == 0]
atl_data = [item for i, item in enumerate(observ_blocks) if i % 2 == 1]

atl_headers
```

```python
import numpy as np

NewFrameDict = {
}

df = pd.DataFrame(NewFrameDict)

for j, block in enumerate(atl_data):
    JD_all = [line.split()[0] for line in block]
    print(JD_all)
    JD = np.array([float(jd) if jd[0] != '!' else float(jd[1:]) for jd in JD_all])
    flags = [True if jd[0] != '!' else False for jd in JD_all]
    mag = np.array([float(line.split()[1]) for line in block])
    # mag_err = np.array([float(line.split()[2]) for line in block])
    ind = np.array([i for i, item in enumerate(block)])
    ind_header = np.array([j for item in block])

    # print(flags)
    
    NewFrameDict = {
        'ind_header' : ind_header,
        'ind' : ind,
        'JD' : JD,
        'mag' : mag,
        # 'mag_err' : mag_err,
        'flags' : flags,
    }
    
    ndf = pd.DataFrame(NewFrameDict)
    
    df = pd.concat([df, ndf], axis = 0, join='outer')

df.info()
df.to_csv(r'/media/olehryz/Elements/progs/univ_codes/2025_2026/04-asteroids/atl.csv', index=False)
```

```python
ind_header = []
MagBand = []
RelPhot = []
RedMag = []
d_sun = []
d_earth = []
PhaseAngle = []
Ecl_Long= []
Ecl_Lat = []
ObsTime = []

for j, item in enumerate(atl_headers):
    ind_header.append(j)
    for line in item:
        if line.startswith('REL'):
            RelPhot.append(line.split()[2])
        if line.startswith('RED'):
            RedMag.append(line.split()[2])
        if line.startswith("ASP"):
            d_sun.append(float(line.split()[2]))
            d_earth.append(float(line.split()[3]))
            PhaseAngle.append(float(line.split()[4]))
            Ecl_Long.append(float(line.split()[5]))
            Ecl_Lat.append(float(line.split()[6]))
        if line.startswith("OBSERVING TIME:"):
            ObsTime.append(line.split('(')[-1][:-1])

NewFrameDict = {
    'index_block' : ind_header,
    'RELPHOT' : RelPhot,
    'REDMAG' : RedMag,
    'DSUN' : d_sun,
    'DEARTH' : d_earth,
    'PA' : PhaseAngle,
    'ELONG' : Ecl_Long,
    'ELAT' : Ecl_Lat,
    'OBS_TIME' : ObsTime,
}
                
df_atl_headers = pd.DataFrame(NewFrameDict)

df_atl_headers.to_csv("atl_headers.csv", index=False)
```
