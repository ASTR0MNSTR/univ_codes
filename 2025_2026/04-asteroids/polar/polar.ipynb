```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/media/olehryz/Elements/progs/univ_codes/2025_2026/04-asteroids/polar/atl_headers.csv')

df
```

```python
df = df.query("ELONG > 225 & ELONG < 270")

df
```

```python
# 2. Define coordinates and colors
# Radius is fixed at 1
r = np.ones(len(df))
# Theta is Elongation (ELONG) converted to radians
theta = np.radians(df['ELONG'])
# Color mapping based on Position Angle (PA)
colors = df['PA']

# 3. Handle OBS-TIME (using index as placeholder if column is missing)
label_col = 'OBS-TIME' if 'OBS-TIME' in df.columns else 'index'
labels = df[label_col] if label_col in df.columns else df.index

# 4. Create the polar plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='polar')

# Plot the points
scatter = ax.scatter(theta, r, c=colors, cmap='viridis', 
                     edgecolors='black', alpha=0.75, s=80)

# Add a colorbar for PA values
cbar = plt.colorbar(scatter, ax=ax, pad=0.1)
cbar.set_label('Position Angle (PA)')

# 5. Add labels (OBS-TIME) to the points
for i, label_text in enumerate(labels):
    ax.annotate(str(label_text), (theta[i], r[i]), 
                fontsize=8, xytext=(5, 5), 
                textcoords='offset points')

# Formatting the plot
ax.set_yticklabels([])  # Hide radial labels since radius is constant
ax.set_ylim(0, 1.2)     # Extend limit slightly to fit labels

# Save the plot
plt.savefig('polar_plot_elong_theta.pdf', bbox_inches='tight', dpi=300)
plt.show()
```
