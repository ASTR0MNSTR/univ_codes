# Simple star model

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Simple star model:
# density profile: rho(r) = rho_c * (1 - (r/R))^n
# n = 1
```

## Linear model (n = 1)

### Density distribution

### Mass distribution

```python
# after integration: m(r) = rho_c * pi * r^3 (4/3 - r/R)

def mass_func(r, rho_c, R):
    return rho_c * np.pi * r**3 * (4/3 - r/R)

R = 1
rho_c = 1
r = np.linspace(0, R, 100)
m = mass_func(r, rho_c, R)

fig, ax = plt.subplots()
ax.plot(r, m)
ax.set_xlabel('r/R')
ax.set_ylabel('m(r)')
plt.show()
```

### Pressure distribution

### Temperature distribution

### Central values

```python
Ms = 2e30 # mass of the star (kg)
Rs = 7e8 # radius of the star (m)
mu = 0.6
G = 6.67430e-11 # gravitational constant (m^3 kg^-1 s^-2)
k_B = 1.380649e-23 # Boltzmann constant (J/K
m_p = 1.6726219e-27 # proton mass (kg)
```

```python
# Central density
rho_c = 3 * Ms / (np.pi * Rs**3) # kg/m^3
print("Central density:", rho_c, "kg/m^3")
```

```python
# Central pressure
P_c = (G * rho_c**2 * Rs**2 * np.pi) * 5/36
print("Central pressure:", P_c, "Pa")
```

```python
# Central temperature
T_c = mu * m_p * G * np.pi * rho_c * (5 * Rs**2 / 36)/ k_B
print("Central temperature:", T_c, "K")
```

### Plots

```python
x = np.arange(0, 1.01, 0.01)
mass = 4*x**3 * (1 - 3*x/4)

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, mass, lw=0.8, color='black')
ax.set_xlabel('r/R; R=1', fontsize=12)
ax.set_ylabel('m(r)/M_central', fontsize=12)

plt.show()
```

```python
temperature = 1 + x - 19*x**2/5 + 9*x**3/5

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, temperature, lw=0.8, color='black')
ax.set_xlabel('r/R; R=1', fontsize=12)
ax.set_ylabel('T(r)/T_central', fontsize=12)

plt.show()
```

```python
pressure = 1 - 24*x**2/5 + 28*x**3/5 - 9*x**4/5

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, pressure, lw=0.8, color='black')
ax.set_xlabel('r/R; R=1', fontsize=12)
ax.set_ylabel('P(r)/P_central', fontsize=12)

plt.show()
```

```python
density = 1 - x

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, density, lw=0.8, color='black')
ax.set_xlabel('r/R; R=1', fontsize=12)
ax.set_ylabel('rho(r)/rho_central', fontsize=12)

plt.show()
```
